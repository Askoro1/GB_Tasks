```python
import re

cnt = 0

for i in range(len(df)):
    found = re.findall(r'@[\S\s]+\[comment=[0-9]+\]', df.iloc[i]["Text"])
    if len(found) > 0:
        print(i, found)
        cnt += 1

cnt
```

Было проверено (см. ячейку ниже), что все тексты перед знаком `'='` - именно из английских или русских букв - так как номера строк, в которых были обнаружены такие подпоследовательности совпадают с номерами строк, в которых были обнаружены подпоследовательности вида `[(text without whitespaces)=(num)]`).**

```python
set1 = set()
set2 = set()

for i in range(len(df)):
    found = re.findall(r'\[[A-Za-zЁёА-я]+=[0-9]+\]', df.iloc[i]["Text"])
    if len(found) > 0:
        set1.add(i)
    found = re.findall(r'\[[\S]+=[0-9]+\]', df.iloc[i]["Text"])
    if len(found) > 0:
        set2.add(i)

set1 == set2
```
