**Разбивать текст романа на главы (каждая глава - отдельный документ) неоптимально. Так как используем кластеризацию для тематического моделирования, то есть каждому документу соотвествует ровно один кластер, будем терять очень много информации о главах, так как логично предположить, что обычно глава книги (особенно учитывая, что в "Преступлении и Наказании" главы достаточно большие и насыщенные) на самом деле может относиться сразу к нескольким темам. Разбивать текст на документы по предложениям - тоже некорректно, так как иногда в предложении может быть сложно выделить тему (а слишком часто обозначать предложения за выбросы - тоже неразумно). Поэтому будем считать за документы абзацы в тексте.**

**Можно заметить, что абзацы разделены двумя переводами строки: `\n\n`. По ним и разобьём текст.**

**Предварительно удалим обозначения частей и глав романа. Такие паттерны имеют следующий вид: `\n\n\n\n\n[text]\n\n\n\n\n`.**

```python
import re

upd_text = re.sub(r'\n\n\n\n\n[A-Za-z0-9]+\n\n\n\n\n', ' ', upd_text) # удаляем так обозначения глав
upd_text = re.sub(r'\n\n\n\n\nЧасть\s[ЁёА-я]+\s', ' ', upd_text) # удаляем так обозначения частей

#upd_text
```

**Разбиваем текст на абзацы.**

```python
dataset = upd_text.split('\n\n')
#dataset
```

****

**Теперь почистим абзацы от пробелов в самом начале и в самом конце, от оставшихся символов перевода строки, от указаний на примечания автора/редакции (в исходном тексте они имеют вид `[num]`, но мы удалили выше скобки, тем не менее из контекста всё равно будет понятно, где то, что нам надо  удалить (см. ниже)), от стоп-слов (это делаем в самом конце), лемматизируем слова (это будет достаточно полезно для тематического моделирования, так как нет большого смысла выделять темы, состоящие из одного и того же слова, но в разных формах).**

```python
import nltk
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
```

```python
!pip install stanza
```

```python
import stanza

stanza.download('ru')

snlp = stanza.Pipeline(lang="ru", use_gpu=True, pos_batch_size=3000, lemma_batch_size=3000)
```

**Пример использования stanza:**

```python
txt = "Андре́й Серге́евич Арша́вин (род. 29 мая 1981[4], Ленинград) — российский футболист, бывший капитан сборной России, заслуженный мастер спорта России (2008)."
doc = snlp(txt)

for sentence in doc.sentences:
    for word in sentence.words:
        print(word.lemma)
```

**Исследуем, какие языки представлены в тексте, кроме русского:**

```python
for text in dataset:
    found = re.findall(r'[A-Za-z]+', text)
    if len(found) > 0:
        print(found)
```

**Из найденного выше следует, что в тексте кроме русского есть немецкий, французский, английский, латынь.**

```python
stop_words = set(stopwords.words('russian') + stopwords.words('english') + stopwords.words('german') + stopwords.words('french')) # латыни конечно в стоп-словах нет :)
```

**Найдём все числа в тексте.**

```python
for text in dataset:
    found = re.findall(r'\d+', text)
    if len(found) > 0:
        print(found)
```

**Числа идут последовательно => можно сделать предположение, что все эти числа - ссылки на примечания, а не имеющие смысл для текста числа. Значит, числа можно удалять.**

**Выполним все описанные выше преобразования текста:**

```python
for text_it in range(len(dataset)):
    dataset[text_it] = dataset[text_it].strip()
    dataset[text_it] = re.sub(r'\d+', ' ', dataset[text_it])
    # dataset[text_it] = dataset[text_it].lower()
    dataset[text_it] = ' '.join(dataset[text_it].split())

dataset
```

```python
from tqdm import tqdm
```

```python
for text_it in tqdm(range(len(dataset))):
    doc = snlp(dataset[text_it])
    lemmatized = []
    for sentence in doc.sentences:
        for word in sentence.words:
            lemmatized += [word.lemma]
    dataset[text_it] = lemmatized
    dataset[text_it] = [word for word in dataset[text_it] if word not in stop_words]

dataset
```

**В тексте есть пустые абзацы (непонятно, как они появились, но всё же). Удалим их.**

```python
dataset_clean = []

for text in dataset:
    if len(text) > 0:
        dataset_clean.append(' '.join(text))

dataset_clean
```

**Приведём все символы к нижнему регистру.**

```python
for text_it in range(len(dataset_clean)):
        dataset_clean[text_it] = dataset_clean[text_it].lower()

dataset_clean
```

**Текст наконец готов. На всякий случай сохраним его.**

```python
import json

with open("dataset.json", "w", encoding='ascii') as f:
    json.dump(dataset_clean, f)
```

**А теперь непосредственно тематическое моделирование.**

**Как будем настраивать пайплайн:**

- **Так как языков в документах - больше одного (не только русский), то будем использовать для генерации эмбеддингов multilingual модель.** 

- **Иногда в текстах бывают какие-либо наименования, которые нет смысла разделять на отдельные слова, так как тогда потеряется смысл наименования (ex: "Соединённые Штаты Америки"). Для этого в BERTopic есть гиперпараметр пайплайна `n_gram_range` - он как раз регулирует, сколько минимум/максимум слов/токенов вместе рассматриваем. Выставим `n_gram_range=(1, 3)`, так как более длинные (больше, чем 3 слова в наименовании) наименования редко встречаются в текстах. UPD: Лучше все же выставить `n_gram_range=(1, 2)`, а то получается что-то совсем странное...**

- **Чтобы не генерировать часто пустые темы или темы-дубликаты, выставим `min_topic_size=20` (таким образом, тему будем создавать, только если в ней будет не менее 20 слов) (дефолтное значение гиперпараметра: `min_topic_size=10`) (применительно только в случае использования HDBSCAN для кластеризации) UPD: неприменительно, не надо это выставлять**



**Попробуем сначала применить KMeans для кластеризации.**

```python
from sklearn.cluster import KMeans

cluster_model = KMeans(n_clusters=30)
```

```python
from bertopic import BERTopic
from bertopic.representation import KeyBERTInspired

representation_model = KeyBERTInspired()

topic_model = BERTopic(language="multilingual", top_n_words=7, n_gram_range=(1, 2), hdbscan_model=cluster_model, representation_model=representation_model, verbose=True)
topics = topic_model.fit_transform(data)
```
