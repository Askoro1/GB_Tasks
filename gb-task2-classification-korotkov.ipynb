**Легко заметить, что в датасете из текстов на русском языке есть английские сочетания вида `[ 0+ английских букв ] +  tokenoid + [ 0+ английских букв ]`. Скорее всего эти сочетания получились при парсинге новостей. Все такие вставки стоит убрать.**

**При этом можно заметить, что иногда (достаточно редко) перед или после `"tokenoid"` есть что-то осмысленное. Например, в датасете есть комбинация `"memestokentokenoid"` - отсюда можно достать слово "memes". Тем не менее, так как такие осмысленные наборы символов - на английском, то проще всего просто их также не рассматривать (тем более что ценной информации в таких фрагментах обычно мало). При этом иногда `"tokenoid"` оказывается сразу за русским текстом (без пробелов), поэтому в регулярном выражении, чистящем эти комбинации символов, надо поставить проверки на удаление именно комбинаций из английских символов.**

```python
df["text"] = df["text"].str.replace(r'[A-Za-z]*tokenoid[A-Za-z]*', ' ', regex=True)
```

В большинстве случаев это число стоит просто удалить из текста (возможно иногда оно вносит в текст какой-то допсмысл, но такое бывает крайне редко (если вообще бывает - мне не удалось найти такие тексты), поэтому было решено просто удалять все вхождения этого числа).**

**Поработаем с пунктуацией. Отделим все "важные" знаки препинания от текста ("важными" знаками препинания будем считать следующие: `'.', ',', '?', '!', ';', ':'`), "неважные" - удалим.**

```python
important_punct = ['.', ',', '?', '!', ';', ':']

for i in important_punct:
    df["text"] = df["text"].str.replace(i, ' ' + i + ' ')
```

```python
unnecessary_punct = ['#', '^', '*', '(', ')', '\\', '/', '\'', '\"', '\n', '\r', '\t', '`']

for i in unnecessary_punct:
    df["text"] = df["text"].str.replace(i, ' ')
```

**Иногда в тексте встречаются эмодзи. Удалим их (так как не всегда понятно, что является эмодзи, будем считать эмодзи все символы, кроме русских букв, английских букв и "важных" знаков препинания).**

```python
df["text"] = df["text"].str.replace(r'[^\sA-Za-z0-9ЁёА-я!\"#\$%&\'\(\)\*\+,-\./:;<=>\?@\[\\\]\^_`{\|}~]', ' ', regex=True)

df.sample(30)
```

- **Попробовать перевести всё предложение (несмотря на то, что оно в большинстве случаев только на русском), чтобы переводчик смог уловить контекст, в котором используется слово или аббревиатура.**

- **Если какое-то слово после перевода предложения не перевелось, попробовать перевести только это слово (скорее всего это какое-то название, а не общепринятое сокращение, которое можно перевести, зная контекст слова).**

- **Если какое-то слово даже после этого не перевелось, его стоит удалить.**

```python
import re
from tqdm import tqdm
```

```python
texts = []

for i in tqdm(range(len(df))):
    text = df.iloc[i]["text"]
    if re.search('[a-zA-Z]', text) is not None:
        text = translator.translate(text, src='en', dest='ru').text
        if re.search('[a-zA-Z]', text) is not None:
            words = text.split()
            for word_it in range(len(words)):
                if re.search('[a-zA-Z]', words[word_it]) is not None:
                    words[word_it] = translator.translate(words[word_it], src='en', dest='ru').text
                    if re.search('[a-zA-Z]', words[word_it]) is not None:
                        words[word_it] = ''
            text = ' '.join(words)
    texts.append(text)

df_clean = pd.Series(texts, name="text")
```

```python
import torch
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
from transformers import AutoTokenizer, AutoModel

model_name = 'Tochka-AI/ruRoPEBert-e5-base-2k'
tokenizer = AutoTokenizer.from_pretrained(model_name)
emb_model = AutoModel.from_pretrained(model_name, trust_remote_code=True, attn_implementation='sdpa').to(device)
```

```python
from tqdm import tqdm
```

```python
HIDDEN_SIZE = 768
BATCH_SIZE = 16
```

```python
def get_embeds(tokenizer, model, texts, device, get_cls_emb=True): 
    # последний аргумент достаёт только общий эмбеддинг для всего текста, если он равен True, иначе кодируем все токены по отдельности (второй кейс, как оказалось, нам не нужен)
    tokens = tokenizer(texts, max_length=2048, padding='max_length', return_tensors="pt")
    if get_cls_emb:
        embeds = np.zeros((len(texts), HIDDEN_SIZE))
    for num in tqdm(range(0, len(tokens["input_ids"]), BATCH_SIZE)):
        if get_cls_emb:
            if num + BATCH_SIZE <= len(texts):
                embed = np.zeros((BATCH_SIZE, HIDDEN_SIZE))
            else:
                embed = np.zeros((len(texts) - num, HIDDEN_SIZE))
        if num + BATCH_SIZE <= len(texts):
            ids, att_mask = tokens["input_ids"][num:(num + BATCH_SIZE)].to(device), tokens["attention_mask"][num:(num + BATCH_SIZE)].to(device)
        else:
            ids, att_mask = tokens["input_ids"][num:].to(device), tokens["attention_mask"][num:].to(device)
        text_tokens = {"input_ids": ids, "attention_mask": att_mask}
        with torch.inference_mode():
            embs = model(**text_tokens)
        if get_cls_emb:
            embed = embs.pooler_output.detach().cpu().numpy()
        if num + BATCH_SIZE <= len(texts):
            embeds[num:(num + BATCH_SIZE), :] = embed
        else:
            embeds[num:, :] = embed
    return embeds
```

```python
emb_train = get_embeds(tokenizer, emb_model, list(X_train), device)
emb_val = get_embeds(tokenizer, emb_model, list(X_val), device)
emb_test = get_embeds(tokenizer, emb_model, list(X_test), device)
```

```python
classes = ['esport', 'martial_arts', 'tennis', 'basketball', 'volleyball',
           'boardgames', 'athletics', 'winter_sport', 'football', 'motosport',
           'autosport', 'extreme', 'hockey']

cls2idx = {}
idx2cls = {}

for num, cls in enumerate(classes):
    cls2idx[cls] = num
    idx2cls[num] = cls
```

```python
def gen_batch(batch, device=device):
    text_arr = []
    label_arr = []
    for text, label in batch:
        text_arr += [text[np.newaxis, :]]
        label_arr += [torch.unsqueeze(label, 0)]
    texts = torch.cat(text_arr, 0)
    labels = torch.cat(label_arr, 0)
    return texts.to(device), labels.to(device) 

def gen_batch_test(batch, device=device):
    text_arr = []
    for i in batch:
        text_arr += [i[0][np.newaxis, :]]
    batch = torch.cat(text_arr, 0)
    return batch.to(device)
```

```python
from torch.utils.data import DataLoader

batch_size = 64

train_loader = DataLoader(dataset_train, batch_size=batch_size,
                        shuffle=True, collate_fn=gen_batch)

val_loader = DataLoader(dataset_val, batch_size=batch_size,
                        shuffle=False, collate_fn=gen_batch)

test_loader = DataLoader(dataset_test, batch_size=batch_size,
                       shuffle=False, collate_fn=gen_batch_test)
```

**Модель:**

```python
from torch import nn
from torch.nn.functional import relu, leaky_relu

class TextClassificationModel(nn.Module):
    def __init__(self, hidden_dim, num_classes):
        super(TextClassificationModel, self).__init__()
        self.fc1 = nn.Linear(hidden_dim, 16 * hidden_dim, dtype=torch.float64)
        self.fc2 = nn.Linear(16 * hidden_dim, num_classes, dtype=torch.float64)
        self.init_weights()

    def init_weights(self):
        torch.nn.init.xavier_uniform(self.fc1.weight)
        self.fc1.bias.data.zero_()
        torch.nn.init.xavier_uniform(self.fc2.weight)
        self.fc2.bias.data.zero_()

    def forward(self, emb):
        return self.fc2(leaky_relu(self.fc1(emb)))
```

**Код обучения:**

```python
def train_epoch(model, optimizer, loader, loss_fn):
    model.train()
    losses = 0

    for embs, labels in tqdm(loader):
        embs = embs.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        # forward pass + backward pass
        logits = model(embs)
        loss = loss_fn(logits, labels)
        loss.backward()
        #torch.nn.utils.clip_grad_norm_(model.parameters(), 0.1)
        optimizer.step()

        losses += loss.item()

    return losses / len(list(loader))


@torch.no_grad()
def eval_epoch(model, loader, loss_fn):
    model.eval()
    losses = 0

    for embs, labels in tqdm(loader):
        embs = embs.to(device)
        labels = labels.to(device)

        # calc loss
        logits = model(embs)
        loss = loss_fn(logits, labels)

        losses += loss.item()

    return losses / len(list(loader))


@torch.no_grad()
def gen_output(model, loader, loader_type="val"):
    model.eval()
    y_pred = []
    if loader_type == "val":
        for embs, _ in tqdm(loader):
            embs = embs.to(device)
            logits = model(embs)
            pred = torch.argmax(logits, dim=1).detach().cpu().numpy()
            y_pred += [pred]
        y_pred = np.concatenate(y_pred)
    else:
        for embs in tqdm(loader):
            embs = embs.to(device)
            logits = model(embs)
            pred = torch.argmax(logits, dim=1).detach().cpu().numpy()
            y_pred += [pred]
        y_pred = np.concatenate(y_pred)
    return y_pred
        

def train(model, optimizer, n_epochs, loss_fn,
          train_loader, val_loader,
          scheduler=None):
    train_loss_log, val_loss_log = [], []

    for epoch in range(1, n_epochs + 1):
        train_loss = train_epoch(model, optimizer, train_loader, loss_fn)
        train_loss_log.append(train_loss)
        
        val_loss = eval_epoch(model, val_loader, loss_fn)
        val_loss_log.append(val_loss)
        
        y_pred = gen_output(model, val_loader)
        
        print(f"Epoch {epoch}")
        print(f" train loss: {train_loss}")
        print(f" val loss: {val_loss}")
        print(f" val accuracy: {accuracy_score(y_val_idx, y_pred)}\n")

        if scheduler is not None:
            if type(scheduler) == torch.optim.lr_scheduler.ReduceLROnPlateau:
                scheduler.step(val_loss)
            else:
                scheduler.step()
            print("learning rate:", *scheduler.get_last_lr())
```

**Запускаем:**

```python
HIDDEN_DIM = 768
NUM_CLASSES = 13

model = TextClassificationModel(HIDDEN_DIM, NUM_CLASSES)

model = model.to(device)

loss_fn = torch.nn.CrossEntropyLoss()

#optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
optimizer = torch.optim.SGD(model.parameters(), lr=0.005, momentum=0.99)

#scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=15, eta_min=5e-5)

n_epochs = 15

train(model, optimizer, n_epochs, loss_fn,
      train_loader, val_loader, scheduler=None)
```
